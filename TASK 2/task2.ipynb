# <font color='red'>Problem Statement: From the given ‘Iris’ dataset, predict the optimum number of clusters and represent it visually >> </font>


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
```

# <font color='red'>ignoring the warnings </font>

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = datasets.load_iris()
iris_df =  pd.DataFrame(iris.data, columns= iris.feature_names)
```

```python
print(iris_df.head())
```

```python
print(iris_df.info())
```

# <font color='red'>Find the optimal number of K >> </font>

```python
X = iris_df.iloc[:,[0,1,2,3]].values
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i,init ='k-means++', max_iter = 300,n_init = 10,random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

# <font color='red'>absorb the elbow point >> </font>

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

# <font color='red'>Obtained values of K is 3 >> </fomt>

```python
kmeans = KMeans(n_clusters = 3,init ='k-means++', max_iter = 300,n_init = 10,random_state = 0)
y_means = kmeans.fit_predict(X)
```

# <font color='red'>Visualizing the Cluster and Plotting centroids of clusters>> </font>

```python
# Visualizing the Cluster
plt.scatter(X[y_means == 0,0],X[y_means == 0,1], s =100, c='red', label ='Iris-setosa')
plt.scatter(X[y_means == 1,0],X[y_means == 1,1], s =100, c='blue', label ='Iris-Versicolour')
plt.scatter(X[y_means == 2,0],X[y_means == 2,1], s =100, c='green', label ='Iris-virginica')
# Plotting centroids of clusters
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s =100, c='yellow', label ='centroids')
plt.legend()
plt.show()
```
